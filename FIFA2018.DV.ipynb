```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('data.csv')

raw.head()
```

```python
# Histogram


bins = [40,50,60,70,80]

plt.hist(raw.Potential, bins=bins, color='#4e42f5')

plt.xticks(bins)

plt.xlabel('Skill Level')
plt.ylabel('No.of Players')

plt.title('Players skills in FIFA 2018')

plt.show()
```

```python
# Pie Chart

left = raw.loc[raw['Preferred Foot'] == 'Left'].count()[0]
right = raw.loc[raw['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#f54b42', '#45f542']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.1f %%')

plt.title('Foots of FIFA players')

plt.show()
```
