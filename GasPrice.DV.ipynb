```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('gas_prices.csv')

data.head()
```

```python
plt.title('Gas Prices in various countries', fontdict={'fontweight':'bold', 'fontsize': 20})

plt.plot(data.Year, data.France, 'b.-', label='France')
plt.plot(data.Year, data.Germany, 'y.-', label='Germany')
plt.plot(data.Year, data.Japan, 'r.-', label='Japan')
plt.plot(data.Year, data.UK, 'g.-', label='UK')

plt.xticks(data.Year[::3].tolist()+[2010])


plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.show()
```
